```python
def filter_prime(nlist):
    for num in nlist:
        isprime=True
        for i in range(2,num//2):
            if num%i==0:
                isprime=False
        if isprime==False :
            nlist.remove(num)
    return nlist
enterlist=input("Enter your numbers:")
nlist=list(map(int,enterlist.split()))
filter_prime(nlist)
```

```python
import math

def regular_polygon_area(n, a):
    return (n * a**2) / (4 * math.tan(math.pi / n))

# Пример: правильный шестиугольник со стороной 5
n = int(input("Введите число сторон: "))
a = float(input("Введите длину стороны: "))

print("Площадь:", regular_polygon_area(n, a))
```
